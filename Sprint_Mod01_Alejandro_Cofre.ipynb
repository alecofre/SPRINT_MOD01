```python
class Producto():
  def __init__(self, id, nombre, descripcion, cantidad:int, precio:int):
    self.id = id
    self.nombre = nombre
    self.descripcion = descripcion
    self.cantidad = cantidad
    self.precio = precio

  def __str__(self):
    return f'{self.id} - {self.nombre} - {self.descripcion} - {self.cantidad} - {self.precio}'

  def get_id(self):
    return self.id

  def set_id(self, nuevo_id):
    self.id = nuevo_id

  def get_nombre(self):
    return self.nombre

  def set_nombre(self, nuevo_nombre):
    self.nombre = nuevo_nombre

  def get_descripcion(self):
    return self.descripcion

  def set_descripcion(self, nuevo_descripcion):
    self.descripcion = nuevo_descripcion

  def get_cantidad(self):
    return self.cantidad

  def set_cantidad(self, nuevo_cantidad):
    self.cantidad = nuevo_cantidad

  def get_precio(self):
    return self.precio

  def set_precio(self, nuevo_precio):
    self.precio = nuevo_precio

class Inventario():
  def __init__(self): # Define el inventario como un diccionario
    self.productos = {}

  def agregar_producto(self, producto): # Método que agrega un producto al inventario, luego de verificar que el ID no existe previamente
    if producto.id not in self.productos.keys():
      self.productos[producto.id] = producto
      msg = 'Producto agregado'
    else:
      msg = 'Id ya existe'
    return msg

  def eliminar_producto(self, id): # Método que elimina un producto del inventario
    try:
      del self.productos[id]
      msg = 'Producto eliminado'
    except KeyError:
      msg = 'Id del producto no existe'
    return msg

  def modificar_producto(self, id, campo, valor): # Método que modifica el valor de una característica del producto
    my_producto = self.buscar_producto(id)
    if  my_producto is not None:
      if  campo == "nombre":
        my_producto.set_nombre(valor)
        msg = 'Nombre de producto actualizado'
      elif campo == "descripcion":
        my_producto.set_descripcion(valor)
        msg = 'Descripcion de producto actualizado'
      elif campo == "cantidad":
        my_producto.set_cantidad(valor)
        msg = 'Cantidad de producto actualizado'
      elif campo == "precio":
        my_producto.set_precio(valor)
        msg = 'Precio de producto actualizado'
      else:
        msg = 'Campo incorrecto'
    else:
       msg = 'No se encontro el  producto con ese ID'
    return msg

  def buscar_producto(self, id): # Método que busca un producto según su ID
    try:
        return self.productos[id]
    except KeyError:
        return None

  def listar_productos(self): # Método que lista todos los productos y además se suman las unidades del inventario y se calcula la valorización de éste
    total_productos = 0
    valor_inventario = 0

    for producto in self.productos:
      print(self.productos[producto])
      total_productos += self.productos[producto].cantidad   # total_productos = total_productos + producto.cantidad
      valor_inventario += self.productos[producto].cantidad * self.productos[producto].precio
    print(f'Inventario contiene {total_productos} productos')
    print(f'Inventario valorizado en $ {valor_inventario}')


def menu(): # Función que controla el flujo del programa
    inventario = Inventario()

    # Se agregan dos productos para tener un inventario inicial
    p1 = Producto('SS01','Avena','Avena sin gluten',100,2000)
    p2 = Producto('SB03','Tofu','Sabor pimenton',50,6500)

    inventario.agregar_producto(p1)
    inventario.agregar_producto(p2)

    while True:
        print("\nMenu:")
        print("1. Agregar producto")
        print("2. Eliminar producto")
        print("3. Actualizar producto")
        print("4. Buscar producto por ID")
        print("5. Listar todos los productos")
        print("0. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1": # Agrega un producto al inventario
            id = input("Ingrese el ID del producto: ")
            nombre = input("Ingrese el nombre del producto: ")
            descripcion = input("Ingrese la descripción del producto: ")
            while True: # Verifica que el ingreso de usuario para la cantidad sea numérico
              try:
                cantidad = int(input("Ingrese la cantidad del producto: "))
                break
              except:
                 print("Error! La cantidad debe ser un número.")
            while True: # Verifica que el ingreso de usuario para el precio sea numérico
              try:
                precio = float(input("Ingrese el precio del producto: "))
                break
              except:
                 print("Error! El precio debe ser un número.")
            producto = Producto(id, nombre, descripcion, cantidad, precio)
            inventario.agregar_producto(producto)

        elif opcion == "2": # Elimina un producto del inventario
            id = input("Ingrese el ID del producto a eliminar: ")
            inventario.eliminar_producto(id)

        elif opcion == "3": # Modifica alguna característica de un producto
            id = input("Ingrese el ID del producto a actualizar: ")
            if inventario.buscar_producto(id) != None:
              while True: # Verifica que el campo a modificarsea válido
                campo =  input("Ingrese el campo a actualizar (Nombre/Descripción/Cantidad/Precio): ").lower()
                if campo in ['nombre','descripción','cantidad','precio']:
                  break
                else:
                  print('Campo no reconocido')
              flag = True
              while flag:
                valor = input("Ingrese el nuevo valor: ")
                if (campo == 'cantidad') or  (campo =='precio'): 
                  try: # Verifica que para cantidad o precio el ingreso del usuario sea numerico
                      valor = int(valor)
                      flag = False
                  except:
                    print("Error! El campo Cantidad y/o Precio deben ser números.")
              print(inventario.modificar_producto(id, campo, valor))
            else:
               print('El producto no existe en inventario')

        elif opcion == "4": # Busca un producto por su ID en el inventario
            id = input("Ingrese el ID del producto a buscar: ")
            producto = inventario.buscar_producto(id)
            if producto:
                print(producto)
            else:
                print("Producto no encontrado.")

        elif opcion == "5": # Lista todos los productos en el inventario
            inventario.listar_productos()
            print('-------')
        elif opcion == "0": # Sale del menu
            break
        else:
            print("Opción inválida. Intente nuevamente.")

# Ejecutar el programa
menu()


```
